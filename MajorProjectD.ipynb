```python
import numpy as np
import pandas as pd
```

```python
df=pd.read_excel('PAML_COVID_Dataset.xlsx')

def tran_symptoms(x):
    if x=='dry cough':
        return 1
    if x=='tiredness':
        return 2
    if x=='fever':
        return 3
def tran_severity(b):
    if b=='Mild':
        return 4
    if b=='Moderate':
        return 5
    if b=='Seviour':
        return 6
#df['Tran_Symptoms']= df['Most common symptoms'].apply(tran_symptoms)
#df['Tran_Severity']= df['Severity'].apply(tran_severity)
df['Most common symptoms']= df['Most common symptoms'].apply(tran_symptoms)
df['Severity']= df['Severity'].apply(tran_severity)

#print(df.dtypes)
df
```

```python
df.dtypes
```

```python

print(df['Most common symptoms'])
x_symptoms=np.array(df['Most common symptoms']).reshape(-1,1)
y_severity=np.array(df['Severity']).reshape(-1,1)
print(x_symptoms)
print(y_severity)
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn=KNeighborsClassifier(1)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x_symptoms,y_severity,test_size=0.2)
```

```python
type(x_train),type(y_train),y_train
```

```python
y_train=y_train.ravel()
y_train
```

```python
knn.fit(x_train,y_train)
```

```python
x_test
```

```python
predict=knn.predict(x_test)
predict
```

```python
score=knn.score(x_test,y_test)
score
```
